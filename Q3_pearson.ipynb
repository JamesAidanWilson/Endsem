```python
from the_quiver import *
from matplotlib import pyplot as graph
from math import exp

data = readtxt("esem_fit1.dat")

X = [data[i][0] for i in range(len(data))] # storing the data
Y = [data[i][1] for i in range(len(data))]

c1,m1 = linear_fit(X,Y)
V1 = [m1*x + c1 for x in X]

graph.plot(X,Y)
graph.plot(X,V1,'-r')
graph.title("LINEAR FIT")
graph.xlabel("Temperature - T")
graph.ylabel("Conductivity - \sigma")
graph.show()

a,p = exponential_fit(X,Y) # a is \sigma_0 and p is \alpha
V2 = [a*exp(p*x) for x in X]

graph.plot(X,Y)
graph.plot(X,V2,'-r')
graph.title("EXPONENTIAL FIT")
graph.xlabel("Temperature - T")
graph.ylabel("Conductivity - \sigma")
graph.show()

r = pearsons_r(X,Y)

print("\sigma_0 = " + str(a) + " \n" + "\ alpha = " + str(p) + "\n Pearson's 'r' is " + str(r))
```
